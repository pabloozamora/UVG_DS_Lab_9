## Emisiones de dióxido de carbono en el mundo de 1990 a 2018

### Gráfica con ajustes por defecto

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de emisiones de CO2
emissions_data = pd.read_csv("data/historical_emissions.csv")

# Selección de países clave para la visualización
selected_countries = ["China", "United States", "European Union (27)", "India"]

# Filtrar los datos para los países seleccionados y años relevantes
emissions_data_filtered = emissions_data[emissions_data['Country'].isin(selected_countries)]

# Transformar los datos para tener los años como columnas
emissions_data_melted = emissions_data_filtered.melt(id_vars=["Country"], 
                                                     value_vars=[str(year) for year in range(1990, 2018 + 1)], 
                                                     var_name="Year", value_name="CO2 Emissions")
emissions_data_melted["Year"] = emissions_data_melted["Year"].astype(int)
# Crear la gráfica de líneas para todos los países sin personalización adicional
plt.figure(figsize=(12, 6))
for country in emissions_data_melted['Country'].unique():
    country_data = emissions_data_melted[emissions_data_melted['Country'] == country]
    plt.plot(country_data['Year'], country_data['CO2 Emissions'], label=country)

# Personalización mínima
plt.legend(title="Country", loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Gráfica optimizada para reducir la carga cognitiva

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de emisiones de CO2
emissions_data = pd.read_csv("data/historical_emissions.csv")

# Selección de países clave para la visualización
selected_countries = ["China", "United States", "European Union (27)", "India"]

# Filtrar los datos para los países seleccionados y años relevantes
emissions_data_filtered = emissions_data[emissions_data['Country'].isin(selected_countries)]

# Transformar los datos para tener los años como columnas
emissions_data_melted = emissions_data_filtered.melt(id_vars=["Country"], 
                                                     value_vars=[str(year) for year in range(1990, 2018 + 1)], 
                                                     var_name="Year", value_name="CO2 Emissions")
emissions_data_melted["Year"] = emissions_data_melted["Year"].astype(int)

# Crear la gráfica de líneas
plt.figure(figsize=(12, 6))
for country in selected_countries:
    country_data = emissions_data_melted[emissions_data_melted['Country'] == country]
    plt.plot(country_data['Year'], country_data['CO2 Emissions'], label=country)
    
    # Agregar una etiqueta de texto al final de cada línea
    plt.text(country_data['Year'].values[-1], 
             country_data['CO2 Emissions'].values[-1], 
             country, 
             verticalalignment='center')

# Personalizar la gráfica
plt.title("Emisiones de CO₂ por País (1990-2018)")
plt.xlabel("Año")
plt.ylabel("Emisiones de CO₂ (MtCO₂e)")

# Eliminar bordes innecesarios
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(False)
plt.tight_layout()
plt.show()
```
