```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'data/depuracion.csv'
depuracion_data = pd.read_csv(file_path)

# Convertir la columna 'Fecha' en ndice para un grfico ms claro
depuracion_data['Fecha'] = pd.to_datetime(depuracion_data['Fecha'], format='%b-%y')
depuracion_data.set_index('Fecha', inplace=True)

# Crear una grfica engaosa manipulando la escala del eje Y
plt.figure(figsize=(10, 5))
plt.plot(depuracion_data.index, depuracion_data['Ventas Reales (unidades)'], marker='o', color='b')

# Manipular el eje Y con un rango muy estrecho
plt.ylim(43000000, 64000000)

# Ttulos y etiquetas
plt.title("Ventas Reales (unidades) - Grfica Engaosa con Escala Manipulada")
plt.xlabel("Fecha")
plt.ylabel("Ventas Reales (unidades)")

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
