### Gr√°fica manipulada

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'data/depuracion.csv'
depuracion_data = pd.read_csv(file_path)

# Convertir la columna 'Fecha' en √≠ndice para un gr√°fico m√°s claro
depuracion_data['Fecha'] = pd.to_datetime(depuracion_data['Fecha'], format='%b-%y')
depuracion_data.set_index('Fecha', inplace=True)

# Crear una gr√°fica enga√±osa manipulando la escala del eje Y
plt.figure(figsize=(10, 5))
plt.plot(depuracion_data.index, depuracion_data['Ventas Reales (unidades)'], marker='o', color='b')

# Manipular el eje Y con un rango muy estrecho
plt.ylim(43000000, 64000000)

# T√≠tulos y etiquetas
plt.title("Ventas Reales (unidades) - Gr√°fica Enga√±osa con Escala Manipulada")
plt.xlabel("Fecha")
plt.ylabel("Ventas Reales (unidades)")

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```

### Gr√°fica acertada

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'data/depuracion.csv'
depuracion_data = pd.read_csv(file_path)

# Convertir la columna 'Fecha' en √≠ndice para un gr√°fico m√°s claro
depuracion_data['Fecha'] = pd.to_datetime(depuracion_data['Fecha'], format='%b-%y')
depuracion_data.set_index('Fecha', inplace=True)

# Crear una gr√°fica enga√±osa manipulando la escala del eje Y
plt.figure(figsize=(10, 3))
plt.plot(depuracion_data.index, depuracion_data['Ventas Reales (unidades)'], marker='o', color='b')

# T√≠tulos y etiquetas
plt.title("Ventas Reales (unidades) - Gr√°fica con Escala Adecuada")
plt.xlabel("Fecha")
plt.ylabel("Ventas Reales (unidades)")

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```
