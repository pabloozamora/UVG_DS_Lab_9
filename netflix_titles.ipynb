## Estadísticas de películas y series de Netflix

### Gráfica con ajustes por defecto

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de Netflix
netflix_data = pd.read_csv("data/netflix_titles.csv")

# Agrupar los datos por año de lanzamiento y tipo
release_counts = netflix_data.groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Crear la gráfica de barras apiladas
release_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.tight_layout()
plt.show()
```

### Gráfica optimizada para reducir la carga cognitiva

```python

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de Netflix
netflix_data = pd.read_csv("data/netflix_titles.csv")

# Crear una nueva columna 'decade' redondeando el año a la década más cercana
netflix_data['decade'] = (netflix_data['release_year'] // 10) * 10

# Agrupar los datos por década y tipo
release_counts = netflix_data.groupby(['decade', 'type']).size().unstack(fill_value=0)

# Crear la gráfica de barras apiladas
release_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#1f77b4', '#ff7f0e'])
plt.title("Producciones de Netflix por Década y Tipo")
plt.xlabel("Década de Lanzamiento")
plt.ylabel("Número de Producciones")
plt.legend(title="Tipo de Contenido", labels=["Películas", "Programas de TV"])  # Traducir leyenda
plt.xticks(rotation=0)
plt.tight_layout()

# Eliminar bordes innecesarios
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
